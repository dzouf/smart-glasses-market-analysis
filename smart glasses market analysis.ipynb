```python
import pandas as pd
import numpy as np

# Dataset 1 : Market Size Forecast (Grand View Research)
years = [2024, 2025, 2026, 2027, 2028, 2029, 2030]
market_size = [1.93, 2.47, 3.15, 4.01, 5.10, 6.49, 8.26]  # Arrondi précis

df_forecast = pd.DataFrame({
    'Year': years,
    'Market Size (USD Billion)': market_size
})

# Calcul YoY Growth (%) automatique
df_forecast['YoY Growth (%)'] = df_forecast['Market Size (USD Billion)'].pct_change() * 100
df_forecast['YoY Growth (%)'] = df_forecast['YoY Growth (%)'].round(1)
df_forecast.loc[0, 'YoY Growth (%)'] = np.nan  # Pas de YoY pour 2024

# Cumulative Growth depuis 2024
df_forecast['Cumulative Growth since 2024 (%)'] = ((df_forecast['Market Size (USD Billion)'] / 1.93 - 1) * 100).round(1)

df_forecast
```
